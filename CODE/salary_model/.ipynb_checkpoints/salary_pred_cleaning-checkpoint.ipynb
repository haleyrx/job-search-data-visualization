### Process raw data scraped from Indeed

```python
df_raw = pd.read_csv('final_salary_data.csv')
df_raw.head()
```

```python
# Drop duplicates
df_cleaned = df_raw.drop_duplicates(subset='link', keep='first')

# Split location into city, state, and country
df_cleaned[['city', 'state']] = df_cleaned['location'].str.split(', ', 1, expand=True)
df_cleaned['country'] = 'United States'
df_cleaned['full_location'] = df_cleaned['city'] + ', ' + df_cleaned['state'] + ', ' + df_cleaned['country']
df_cleaned.head()
```

```python
# Drop rows where there is no salary data
df_cleaned = df_cleaned.dropna(subset=['salary']) 
df_cleaned.info()
```

### Convert salary to yearly 

Since the salary information is in different formats, we need to normalize the data to single float with the annual salary for each listing

* Salary type is labeled and the salary column is parsed into an array [min_salary, max_salary]
* The midpoint of the range is then converted to an annual salary based on the identified salary type

```python
# Get salary type
def get_salary_type(x):
    if "year" in str(x):
        salary_type  = "Yearly"
    elif "month" in str(x):
        salary_type  = "Monthly"
    elif "week" in str(x):
        salary_type  = "Weekly"
    elif "day" in str(x):
        salary_type = "Daily"
    elif "hour" in str(x):
        salary_type  = "Hourly"
    else:
        salary_type = ""
    
    return salary_type
```

```python
df_cleaned['salary_type'] = df_cleaned['salary'].apply(get_salary_type)
df_cleaned.groupby('salary_type').salary.count()
```

```python
# Parse salary for numeric data
def get_salary(sal_string):
    pattern = re.compile('\$[0-9]*[.,]?[0-9]*')
    parsed = pattern.findall(sal_string)
    cleaned = []
    for num in parsed:
        num = num.replace('$', '')
        num = num.replace(',', '')
        cleaned.append(float(num))
    return cleaned

df_cleaned['salary_numeric'] = df_cleaned['salary'].apply(get_salary)
df_cleaned.head(n=10)
```

```python
# Calculate annual salary
def get_adjusted_salary(salary_type, salary_array):
    adj_salaries = []
    adj_lookup = {'Yearly': 1, 'Monthly': 12, 'Weekly': 52, 'Daily': 5*52, 'Hourly': 40*52}
    for t, s in zip(salary_type, salary_array):
        avg_salary = sum(s) / len(s)
        adj_salaries.append(avg_salary * adj_lookup[t])
    return adj_salaries

df_cleaned['normalized_salary'] = get_adjusted_salary(df_cleaned['salary_type'], df_cleaned['salary_numeric'])
df_cleaned.head(n=10)
```

### Get cost of living data
Data is sourced from https://www.numbeo.com/cost-of-living/rankings.jsp. We use the cost of living index as a feature in our predictions. COI data is only available for a subset of US cities. When available, the city COI is mapped to the salary data. To fill in missing data, we average the available indices by state and map the COI using the state only. 

```python
# World COI data by city 
tables = pd.read_html('https://www.numbeo.com/cost-of-living/rankings.jsp')

df_col = tables[1]
df_col = df_col.drop(columns=["Rank"])
df_col.head()
```

```python
# US COI data by city 
tables_2 = pd.read_html('https://www.numbeo.com/cost-of-living/country_result.jsp?country=United+States')
df_us_col = tables_2[4]
df_us_col = df_us_col.drop(columns=["Rank"])
df_us_col.head()
```

```python
df_us_col[['city', 'state']] = df_us_col['City'].str.split(', ', 1, expand=True)
df_us_state = df_us_col[['city', 'state', 'Cost of Living Index']]
df_avg_col = df_us_state.groupby('state')['Cost of Living Index'].mean()
df_avg_col = df_avg_col.to_frame()
df_avg_col.reset_index(inplace=True)
df_avg_col.head()
```

```python
# Merge COI and salary data by city

# df_col['location_array'] = df_col['City'].apply(lambda x: x.split(", "))
# df_col.head()
df_merged = df_cleaned.merge(df_col[['City', 'Cost of Living Index']], left_on='full_location', right_on='City', how = 'left')
df_merged.drop(columns=['City'], inplace=True)
```

```python
# If NaN, lookup COI by state
df_merged['Cost of Living Index'] = np.where(df_merged['Cost of Living Index'].isnull(), df_merged['state'].map(df_avg_col.set_index('state')['Cost of Living Index']), df_merged['Cost of Living Index'])
df_merged.head()
```

```python
df_merged = df_merged.rename(columns={'Cost of Living Index': 'living_index'})
```

### Parse titles and levels

```python
def parse_title(title):
    title = title.lower()
    
    title_list = ['scientist',
                  'analyst',
                  'engineer',
                  'manager', 
                  'sales']
    
    for val in title_list:
        if val in title:
            return val
   
    return 'other'

def parse_level(title):
    title = title.lower()
    
    replacement_dict = {
        'sr': 'senior',
        'staff': 'senior',
        'jr': 'junior',
        'vp': 'manager',
        'lead': 'manager',
        'manager': 'manager',
        'chief': 'manager',
        ' i ': ' junior ',
        ' iii ': ' senior ',
        ' 3 ' : ' senior ',
        ' 1 ' :' junior ',
        'principal': 'senior',
        'director': 'manager',
        'temporary': 'intern',
        'part-time': 'intern',
        'contract': 'intern',

    }
    
    for key,value in replacement_dict.items():
        title = title.replace(key,value)
    
    level_list = ['senior',
                  'junior',
                  'intern',
                  'manager']
    
    for level in level_list:
        if level in title:
            return level
        
    return 'n/a'
    
```

```python
df_merged['title_parsed'] = df_merged['title'].apply(parse_title)
df_merged['level'] = df_merged['title'].apply(parse_level)
```

### Bucket salary into bins

```python
def bucket_salary(x):
    if x > 140000:
        output = '>140K'
    elif x > 100000:
        output = '100K - 140K'
    elif x > 80000:
        output = '80K - 100K'
    elif x > 60000:
        output = '60K - 80K'
    else:
        output = '<60K'
        
    return output

df_merged['target'] = df_merged['normalized_salary'].apply(bucket_salary)
```

### Model 1: TF-IDF Vectorizer Random Forest Model

```python
# Prepare pipeline

categorical_features = ['company', 'title_parsed','level']
text_features = 'description'
numeric_features = ['living_index']

categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(fill_value = 'missing', strategy = 'constant' )),
    ('onehot', OneHotEncoder(handle_unknown ='ignore'))])

text_transformer = TfidfVectorizer(max_features = 10000)

numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())])

preprocessor = ColumnTransformer(
    transformers=[
        ('text', text_transformer, text_features),
        ('cat', categorical_transformer, categorical_features),
        ('num', numeric_transformer, numeric_features)])

clf1 = Pipeline(steps=[('preprocessor', preprocessor),
                      ('classifier', RandomForestClassifier(n_estimators = 300, max_depth = 50))])
```

```python
# Train and run model 

X = df_merged[categorical_features + [text_features] + numeric_features]
y = df_merged['target']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0, stratify = y)

clf1.fit(X_train, y_train)
y_pred = clf1.predict(X_test)

print(classification_report(y_test,y_pred))
```

### Model 3: TF-IDF Vectorizer SGD

```python
# Prepare pipeline

categorical_features = ['salary_type','company','title']
text_features = 'description'
numeric_features = ['living_index']

categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(fill_value = 'missing', strategy = 'constant' )),
    ('onehot', OneHotEncoder(handle_unknown ='ignore'))])

text_transformer = TfidfVectorizer(max_features = 10000)

numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())])

preprocessor = ColumnTransformer(
    transformers=[
        ('text', text_transformer, text_features),
        ('cat', categorical_transformer, categorical_features),
        ('num', numeric_transformer, numeric_features)])

clf = Pipeline(steps=[('preprocessor', preprocessor),
                      ('classifier', SGDClassifier(alpha=0.001, max_iter=500, n_jobs=3))])

# Train and run model 

X = df_merged[categorical_features + [text_features] + numeric_features]
y = df_merged['target']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0, stratify = y)

parameters = {
    'classifier__alpha': (0.0001, 0.001),
    'classifier__max_iter': (100, 200, 500) 
}

clf_grid = GridSearchCV(clf, parameters)
  

clf_grid.fit(X_train, y_train)
best_parameters = clf_grid.best_estimator_.get_params()
```

```python
for param_name in sorted(parameters.keys()):
    print("\t%s: %r" % (param_name, best_parameters[param_name]))
```

```python
y_pred = clf_grid.predict(X_test)

print(classification_report(y_test,y_pred))
```

```python
stp = stopwords.words('english')
#stp += list(string.punctuation)
stp += ['llc', 'www','http','com', 'pm','am','eg','e','g','ie']

def clean_word(descriptions, stopwords):
    tokenized = []
    pattern = "([a-zA-Z]+(?:'[a-z]+)?)"

    for d in descriptions:
        tokens = nltk.regexp_tokenize(d, pattern)
        words = [word for word in tokens if word not in stopwords]
        tokenized.append((' ').join(words))
    return pd.Series(tokenized)
```
